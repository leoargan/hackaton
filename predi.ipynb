Feature Engineering
Ahora vamos a evaluar el rendimiento 
Para demostrar cómo funciona esto, considere los datos, que indican si Street hay acceso a la propiedad. 

```python
#Transformación 
print ("Original: \n")
print (train.Street.value_counts(), "\n")
```

En la columna street, los valores únicos son Pavey Grvl, que describen el tipo de camino de acceso a la propiedad. En el conjunto de entrenamiento, solo 5 viviendas tienen acceso de grava. Nuestro modelo necesita datos numéricos, por lo que utilizaremos la codificación one-hot para transformar los datos en una columna booleana.

Creamos una nueva columna llamada enc_street. El método pd.get_dummies() realizará esto por nosotros.

Lo hacemos tanto en el archivo de testeo como en el de entrenamiento 

```python
train['enc_street'] = pd.get_dummies(train.Street, drop_first=True)
test['enc_street'] = pd.get_dummies(train.Street, drop_first=True)
```

```python
print ('Encoded: \n')
print (train.enc_street.value_counts())
```

Ya viendo que todo funciona vamos a diseñar otra función con SaleCondition y una tabla dinámica como utilizamos anteriormente para OverallQual

```python
condition_pivot = train.pivot_table(index='SaleCondition', values='SalePrice', aggfunc=np.median)
condition_pivot.plot(kind='bar', color='blue')
plt.xlabel('Sale Condition')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

Para crear esta función, al tener Partial un precio superior de venta al resto lo vamos a codificar para que sea 1 y el resto 0

```python
def encode(x):
     return 1 if x == 'Partial' else 0
train['enc_condition'] = train.SaleCondition.apply(encode)
test['enc_condition'] = test.SaleCondition.apply(encode)
```

Lo graficamos

```python
condition_pivot = train.pivot_table(index='enc_condition', values='SalePrice', aggfunc=np.median)
condition_pivot.plot(kind='bar', color='blue')
plt.xlabel('Encoded Sale Condition')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

Para comenzar con el modelado vamos a completar los valores faltantes con un interpolado

```python
data = train.select_dtypes(include=[np.number]).interpolate().dropna()
```

```python
sum(data.isnull().sum() != 0)
```

Construyo modelo lineal, sin SalePrice ni Id y creamos un conjunto de entrenamiento y otro de espera 

```python
y = np.log(train.SalePrice)
X = data.drop(['SalePrice', 'Id'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                          X, y, random_state=42, test_size=.33)
```

Instanciamos el modelo de Regresión lineal 

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
#Performance
print ("R^2 is: \n", model.score(X_test, y_test))
```

```python
#Predicción 
predictions = model.predict(X_test)
```

Predicción por RMSLE

```python
from sklearn.metrics import mean_squared_error

lin_mse = mean_squared_error(y_test, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
# RMSLE
from sklearn.metrics import mean_squared_log_error
mean_squared_log_error(y_test, predictions, squared=False)
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=.7, color='b') #alpha helps to show overlapping data
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model')
plt.show()
```

```python
#Improve the model 
for i in range (-2, 3):
    alpha = 10**i
    rm = linear_model.Ridge(alpha=alpha)
    ridge_model = rm.fit(X_train, y_train)
    preds_ridge = ridge_model.predict(X_test)

    plt.scatter(preds_ridge, actual_values, alpha=.75, color='b')
    plt.xlabel('Predicted Price')
    plt.ylabel('Actual Price')
    plt.title('Ridge Regularization with alpha = {}'.format(alpha))
    overlay = 'R^2 is: {}\nRMSE is: {}'.format(
                    ridge_model.score(X_test, y_test),
                    mean_squared_error(y_test, preds_ridge))
    
    plt.show()
```
